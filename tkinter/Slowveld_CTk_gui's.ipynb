### Slowveld 

Pricing calculator  
All in one class.  
  
Creates a customtkinter GUI tool to calculate the different pricing tiers associated to Slowveld candles 

```python
import customtkinter

class SlowCalc(customtkinter.CTk):
        '''Slowveld Calculations: 1. Candle Price,'''
        CANDLE_DICT = {
                "100":[{"price": 190,
                        "quantity": 1,
                        "sell_type":"Retail"},
                        {"price": 150,
                        "quantity": 10,
                        "sell_type":"Early Adopter"},
                        {"price": 140,
                        "quantity": 30,
                        "sell_type":"Building the brand"},
                        {"price": 130,
                        "quantity": 50,
                        "sell_type":"Loving the brand"}],
                "200":[{"price": 290,
                        "quantity": 1,
                        "sell_type":"Retail"},
                        {"price": 210,
                        "quantity": 10,
                        "sell_type":"Early Adopter"},
                        {"price": 200,
                        "quantity": 30,
                        "sell_type":"Building the brand"},
                        {"price": 190,
                        "quantity": 50,
                        "sell_type":"Loving the brand"}],
                "500":[{"price": 420,
                        "quantity": 1,
                        "sell_type":"Retail"},
                        {"price": 270,
                        "quantity": 10,
                        "sell_type":"Early Adopter"},
                        {"price": 260,
                        "quantity": 30,
                        "sell_type":"Building the brand"},
                        {"price": 250,
                        "quantity": 50,
                        "sell_type":"Loving the brand"}]
                        }

        CANDLE_RANGE={'500':
                        ["Beneath the Baobab",
                         "In the Riverbed",
                         "Marula Bloom",
                         ""],
                      '200':[
                         "Firewood",
                         "Under the Stars",
                         "Among Elephants",
                         "Last Minute Leopard"],
                      '100':[
                         "Firewood",
                         "Under the Stars",
                         "Among Elephants",
                         "Last Minute Leopard"],
                        }

        DISCOUNT_MOQ = ['1','10','30','50']

        def __init__(self):
                super().__init__()

                self.geometry("350x450")
                self.title("Slowveld Pricing Tool")

                # create 2x2 grid system
                # self.grid_rowconfigure(0, weight=1)
                # self.grid_columnconfigure((0, 2), weight=1)

                # Candle Size ComboBox - wiith dropdown
                self.candle_size = customtkinter.CTkComboBox(master=self,
                                                                 values=self.candle_size_values(),
                                                                 )
                self.candle_size_label = customtkinter.CTkLabel(master=self,
                                                                text = "Candle Range")
                self.candle_size_label.grid(row=0,
                                         column=0,
                                         padx=20,
                                         pady=(20, 0),
                                         sticky="nw")
                self.candle_size.grid(row=0,
                                         column=1,
                                         padx=20,
                                         pady=(20, 0),
                                         sticky="nw")

                # Quantity of Candles
                self.candle_quantity = customtkinter.CTkComboBox(master=self,
                                                                    values=self.DISCOUNT_MOQ)
                self.candle_quantity_label = customtkinter.CTkLabel(master = self,
                                                                        text="Number of Candles")
                self.candle_quantity_label.grid(row=1,
                                         column=0,
                                         padx=20,
                                         pady=(20, 0),
                                         sticky="nw")
                self.candle_quantity.grid(row=1,
                                         column=1,
                                         padx=20,
                                         pady=(20, 0),
                                         sticky="nw")
                
                # CheckBox - sum candle cost
                self.sum_candles = customtkinter.CTkCheckBox(master=self,
                                                                text='Sum candles',
                                                                )
                self.sum_candles.grid(row=2,
                                         column=0,
                                         columnspan=1,
                                         padx=20,
                                         pady=(20, 0),
                                         sticky="ne")

                # Textbox to show calulation results
                self.textbox = customtkinter.CTkTextbox(master=self,)
                self.textbox.grid(row=3,
                                   column=0,
                                   columnspan = 2,
                                   padx=20,
                                   pady=(20, 0),
                                   sticky="ew")


                self.button = customtkinter.CTkButton(master=self,
                                                        command=lambda: self.candle_price(self.candle_size.get(),int(self.candle_quantity.get()),self.sum_candles.get()),
                                                        text="Calculate")
                self.button.grid(row=4,
                                 column=0,
                                 columnspan=2,
                                 padx=20,
                                 pady=20,
                                 sticky="ew")


        def candle_size_values(self):
                return [candle_size for candle_size in self.CANDLE_DICT]

        def candle_plural(self,quantity):
                if quantity > 1:
                        return "candles"
                else:
                        return "candle"

        def candle_price(self,candle_range_size,quantity: int, sum_up=0):
                '''Calculate the price of the candle'''
                cd = self.CANDLE_DICT[str(candle_range_size)]
                
                if quantity < cd[1]["quantity"]:
                        price = cd[0]["price"]
                        category = cd[0]["sell_type"]
                elif cd[1]["quantity"] <= quantity < cd[2]["quantity"]:
                        price = cd[1]["price"]
                        category = cd[1]["sell_type"]
                elif cd[2]["quantity"] <= quantity < cd[3]["quantity"]:
                        price = cd[2]["price"]
                        category = cd[2]["sell_type"]
                elif cd[3]["quantity"] <= quantity:
                        price = cd[3]["price"]
                        category = cd[3]["sell_type"]


                if sum_up==1:
                        return self.textbox.insert("insert", f"{quantity} {self.candle_plural(quantity)} in {candle_range_size}g range, cost R{str(float(price)*quantity)} in total"  + "\n")
                elif sum_up==0: 
                        return self.textbox.insert("insert",f"{quantity} {self.candle_plural(quantity)} in {candle_range_size}g range, cost R{str(float(price))} per unit"  + "\n")
                        


if __name__ == "__main__":

    sc = SlowCalc()
    sc.mainloop()
```
