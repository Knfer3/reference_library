# Conversion measures

### Length

##### Millimetres

```python
# MM to CM
def mm_to_cm(mm):
    return float(mm/10)

# MM to M
def mm_to_m(mm):
    return float(mm/1000)

# MM to KM
def mm_to_km(mm):
    return float(mm/1000000)

# MM to Inches
def mm_to_inch(mm):
    return float(mm*0.0393701)

# MM to Yards
def mm_to_yard(mm):
    return float(mm*0.00109361)


# MM to Miles
def mm_to_mile(mm):
    return float(mm*6.2137e-7)
    
```

##### Centimeters

```python
# CM to MM
def cm_to_mm(cm):
    return float(cm*10)


# CM to M
def cm_to_m(cm):
    return float(cm/100)

# CM to KM
def cm_to_km(cm):
    return float(cm/100000)

# CM to Inch
def cm_to_inch(cm):
    return float(cm*0.3937)

# CM to Yards
def cm_to_yard(cm):
    return float(cm*0.010936111996938)
    
# CM to Miles
def cm_to_mile(cm):
    return float(cm*6.2137e-6)
```

##### Meters

```python

# M to MM
def m_to_mm(m):
    return float(m*1000)

# M to CM
def m_to_cm(m):
    return float(m*100)

# M to KM
def m_to_km(m):
    return float(m/1000)

# M to inches
def m_to_inch(m):
    return float(m*39.37006655984252)
# M to Yrds
def m_to_yard(m):
    return float(m*1.0936)

# M to Miles
def m_to_mile(m):
    return float(m*0.000621371)
```

```python
class Convert:
    '''To perform conversions [.length(),.weight(),.volume()]'''

    
    
    def length(self,amount,convert_from,convert_to):
        '''Length conversions: (amount, convert_from,convert_to)'''
        
        if convert_from == "mm":
            if convert_to == "cm":
                result = float(amount/10)
            elif convert_to == "m":
                result = float(amount/1000)
            elif convert_to == "km":
                result = float(amount/1000000)
            elif convert_to == "inch":
                result = float(amount*0.0393701)
            elif convert_to == "yard":
                result = float(amount*0.00109361)
            elif convert_to == "mile":
                result = float(amount*6.2137e-7)
            else:
                return f'INVALID CONVERT TO LENGTH- {convert_to}'
        elif convert_from == "cm":
            if convert_to == "mm":
                result = float(amount*10)
            elif convert_to == "m":
                result = float(amount/100)
            elif convert_to == "km":
                result = float(amount/100000)
            elif convert_to == "inch":
                result = float(amount*0.3937)
            elif convert_to == "yard":
                result = float(amount*0.010936111996938)
            elif convert_to == "mile":
                result = float(amount*6.2137e-6)
            else:
                return f'INVALID CONVERT TO LENGTH- {convert_to}'

        elif convert_from == "m":
            if convert_to == "mm":
                result = float(amount*1000)
            elif convert_to == "cm":
                result = float(amount*100)
            elif convert_to == "km":
                result = float(amount/1000)
            elif convert_to == "inch":
                result = float(amount*39.37006655984252)
            elif convert_to == "yard":
                result = float(amount*1.0936)
            elif convert_to == "mile":
                result = float(amount*0.000621371)
            else:
                return f'INVALID CONVERT TO LENGTH- {convert_to}'

        elif convert_from == "km":
            if convert_to == "mm":
                result = float(amount*1000000)
            elif convert_to == "cm":
                result = float(amount*100000)
            elif convert_to == "m":
                result = float(amount*1000)
            elif convert_to == "inch":
                result = float(amount*39370.1)
            elif convert_to == "yard":
                result = float(amount*1093.613888889)
            elif convert_to == "mile":
                result = float(amount*0.621371)
            else:
                return f'INVALID CONVERT TO LENGTH- {convert_to}'

        elif convert_from == "inch":
            if convert_to == "mm":
                result = float(amount*25.3999512574)
            elif convert_to == "cm":
                result = float(amount*2.5399951257399999704)
            elif convert_to == "m":
                result = float(amount*0.0254)
            elif convert_to == "km":
                result = float(amount*2.54e-5)
            elif convert_to == "yard":
                result = float(amount*0.0277778)
            elif convert_to == "mile":
                result = float(amount*1.5783e-5)
            else:
                return f'INVALID CONVERT TO LENGTH- {convert_to}'

        elif convert_from == "yard":
            if convert_to == "mm":
                result = float(amount*914.4)
            elif convert_to == "cm":
                result = float(amount*91.44)
            elif convert_to == "m":
                result = float(amount*0.9144)
            elif convert_to == "km":
                result = float(amount*0.0009144)
            elif convert_to == "inch":
                result = float(amount*36)
            elif convert_to == "mile":
                result = float(amount*0.000568182)
            else:
                return f'INVALID CONVERT TO LENGTH- {convert_to}'

        elif convert_from == "mile":
            if convert_to == "mm":
                result = float(amount*1.609e+6)
            elif convert_to == "cm":
                result = float(amount*160899.999998412)
            elif convert_to == "m":
                result = float(amount*1609.3)
            elif convert_to == "km":
                result = float(amount*1.6093)
            elif convert_to == "inch":
                result = float(amount*63360)
            elif convert_to == "yard":
                result = float(amount*1760)
            else:
                return f'INVALID CONVERT TO LENGTH- {convert_to}'

        else:
            return f'INVALID CONVERT FROM LENGTH- {convert_from}'
        return f'{float(amount)}{convert_from} = {result}{convert_to}'

    def weight(self,amount,convert_from,convert_to):
        '''weight conversion (amount, convert_from,convert_to)'''
        if convert_from == "mg":
            if convert_to == "g":
                result = float(amount/1000)
            elif convert_to == "kg":
                result = float(amount/1e-6)
            elif convert_to == "t":
                result = float(amount/1e-9)
            elif convert_to == "oz":
                result =float(amount*3.5274e-5)
            elif convert_to == "lb":
                result =float(amount*2.2046e-6)
            else:
                return f'INVALID CONVERT FROM: {convert_from}'
        elif convert_from == "g":
            if convert_to == "mg":
                result = float(amount*1000)
            elif convert_to == "kg":
                result = float(amount/1000)
            elif convert_to == "t":
                result = float(amount/1000000)
            elif convert_to == "oz":
                result = float(amount/0.035274)
            elif convert_to == "lb":
                result = float(amount/0.00220462)
            else:
                return f'INVALID CONVERT FROM: {convert_from}'
        elif convert_from == "kg":
            if convert_to == "mg":
                result = float(amount*1000000)
            elif convert_to == "g":
                result = float(amount*1000)
            elif convert_to == "t":
                result = float(amount/1000)
            elif convert_to == "oz":
                float(amount/35.274)
            elif convert_to == "lb":
                float(amount/2.20462)
            else:
                return f'INVALID CONVERT FROM: {convert_from}'
        elif convert_from == "t":
            if convert_to == "mg":
                result = float(amount*1000000000)
            elif convert_to == "g":
                result = float(amount/1000000)
            elif convert_to == "kg":
                result = float(amount/1000)
            elif convert_to == "oz":
                result = float(amount/35274)
            elif convert_to == "lb":
                result = float(amount/2204.62)
            else:
                return f'INVALID CONVERT FROM: {convert_from}'
        elif convert_from == "oz":
            if convert_to == "mg":
                result = float(amount*28349.4894089122)
            elif convert_to == "g":
                result = float(amount*28.3495)
            elif convert_to == "kg":
                result = float(amount*0.0283495)
            elif convert_to == "t":
                result = float(amount*2.835e-5) 
            elif convert_to == "lb":
                result = float(amount*0.0625)
            else:
                return f'INVALID CONVERT FROM: {convert_from}'
        elif convert_from == "lb":
            if convert_to == "mg":
                result = float(amount*453592)
            elif convert_to == "g":
                result = float(amount*453.592)
            elif convert_to == "kg":
                result = float(amount*0.453592)
            elif convert_to == "t":
                result = float(amount*0.000453592)
            elif convert_to == "oz":
                result = float(amount*16)
            else:
                return f'INVALID CONVERT FROM: {convert_from}'
        else:
            return f'INVALID CONVERT TO: {convert_to}'
            # Change later: return float only
        return f"{amount}{convert_from} = {result}{convert_to}"
    
    def volume(self,amount,convert_from, convert_to):
        '''Volume conversion (amount, convert_from,convert_to)'''

        if convert_from == "ml":
            if convert_to == "l":
                result = float(amount*0.001)
            elif convert_to == "teaspoon":
                result = float(amount*0.168936)
            elif convert_to == "tablespoon":
                result = float(amount*0.0563121)
            elif convert_to == "drops":
                result = float(amount*20)
            else:
                return f'INVALID CONVERT_TO VOLUME: {convert_to}'
        elif convert_from == "l":
            if convert_to == "ml":
                result = float(amount*1000)
            elif convert_to == "teaspoon":
                result = float(amount*168.936)
            elif convert_to == "tablespoon":
                result = float(amount*56.3121)
            elif convert_to == "drops":
                result = float(amount*20000)
            else:
                return f'INVALID CONVERT_TO VOLUME: {convert_to}'
        elif convert_from == "teaspoon":
            if convert_to == "ml":
                result = float(amount*5.91939)
            elif convert_to == "l":
                result = float(amount*0.00591939)
            elif convert_to == "tablespoon":
                result = float(amount*0.33333)
            elif convert_to == "drops":
                result = float(amount*118.388)
            else:
                return f'INVALID CONVERT_TO VOLUME: {convert_to}'
        elif convert_from == "tablespoon":
            if convert_to == "ml":
                result = float(amount*17.7582)
            elif convert_to == "l":
                result = float(amount*0.0177582)
            elif convert_to == "teaspoon":
                result = float(amount*3)
            elif convert_to == "drops":
                result = float(amount*355.163)
            else:
                return f'INVALID CONVERT_TO VOLUME: {convert_to}'
        elif convert_from == "drops":
            if convert_to == "ml":
                result = float(amount*0.05)
            elif convert_to == "l":
                result = float(amount*5e-5)
            elif convert_to == "teaspoon":
                result = float(amount*0.00844682)
            elif convert_to == "tablespoon":
                result = float(amount*0.00281561)
            else:
                return f'INVALID CONVERT_TO VOLUME: {convert_to}'
        else:
            return f'INVALID CONVERT_FROM VOLUME: {convert_from}'
        return f"{amount}{convert_from} : {result}{convert_to}"

```
