# Conversion measures

### Length

##### Millimetres

```python
# MM to CM
def mm_to_cm(mm):
    return float(mm/10)

# MM to M
def mm_to_m(mm):
    return float(mm/1000)

# MM to KM
def mm_to_km(mm):
    return float(mm/1000000)

# MM to Inches
def mm_to_inch(mm):
    return float(mm*0.0393701)

# MM to Yards
def mm_to_yard(mm):
    return float(mm*0.00109361)


# MM to Miles
def mm_to_mile(mm):
    return float(mm*6.2137e-7)
    
```

##### Centimeters

```python
# CM to MM
def cm_to_mm(cm):
    return float(cm*10)


# CM to M
def cm_to_m(cm):
    return float(cm/100)

# CM to KM
def cm_to_km(cm):
    return float(cm/100000)

# CM to Inch
def cm_to_inch(cm):
    return float(cm*0.3937)

# CM to Yards
def cm_to_yard(cm):
    return float(cm*0.010936111996938)
    
# CM to Miles
def cm_to_mile(cm):
    return float(cm*6.2137e-6)
```

##### Meters

```python

# M to MM
def m_to_mm(m):
    return float(m*1000)

# M to CM
def m_to_cm(m):
    return float(m*100)

# M to KM
def m_to_km(m):
    return float(m/1000)

# M to inches
def m_to_inch(m):
    return float(m*39.37006655984252)
# M to Yrds
def m_to_yard(m):
    return float(m*1.0936)

# M to Miles
def m_to_mile(m):
    return float(m*0.000621371)
```

##### Kilometers

```python
# KM to MM
def km_to_mm(km):
    return float(km*1000000)

# KM to CM
def km_to_cm(km):
    return float(km*100000)

# KM to M
def km_to_m(km):
    return float(km*1000)

# KM to inches
def km_to_inch(km):
    return float(km*39370.1)

# KM to yards
def km_to_yard(km):
    return float(km*1093.613888889)

# KM to Mile
def km_to_mile(km):
    return float(km*0.621371)
```

##### Inches

```python
# inch to MM
def inch_to_mm(inch):
    return (inch*25.3999512574)

# inch to CM
def inch_to_cm(inch):
    return (inch*2.5399951257399999704)

# inch to M
def inch_to_m(inch):
    return (inch*0.0254)

# inch to KM
def inch_to_km(inch):
    return (inch*2.54e-5)

# inch to yards
def inch_to_yard(inch):
    return (inch*0.0277778)

# inch to MM
def inch_to_mile(inch):
    return (inch*1.5783e-5)
```

##### Yards

```python
# Yards to MM
def yard_to_mm(yard):
    return float(yard*914.4)

# Yards to CM
def yard_to_cm(yard):
    return float(yard*91.44)

# Yrds to M
def yard_to_m(yard):
    return float(yard*0.9144)

# Yards to KM
def yard_to_km(yard):
    return float(yard*0.0009144)

# Yrds to inches
def yard_to_inch(yard):
    return float(yard*36)

# Yards to Miles
def yard_to_mile(yard):
    return float(yard*0.000568182)
```

##### Miles

```python
# Mile to MM
def mile_to_mm(mile):
    return float(mile*1.609e+6)

# Mile to CM
def mile_to_cm(mile):
    return float(mile*160899.999998412)

# Mile to M
def mile_to_m(mile):
    return float(mile*1609.3)

# Mile to KM
def mile_to_km(mile):
    return float(mile*1.6093)

# Mile to inches
def mile_to_inch(mile):
    return float(mile*63360)

# Mile to Yard
def mile_to_yard(mile):
    return float(mile*1760)
```

```python
class Convert:
    '''To do length conversions'''
    def __init__(self,length1,length1_name,length2_name):
        self.length1 = length1
        self.length1_name = length1_name
        self.length2_name = length2_name

    def summary(self,length2):
        '''string output of the conversion'''
        return f'{float(self.length1)}{self.length1_name} = {length2}{self.length2_name}'
    
    
    def length(self):
        '''Calls the conversion function for each metric'''
        if self.length1_name == "mm":
            if self.length2_name == "cm":
                result = mm_to_cm(self.length1)
            elif self.length2_name == "m":
                result = mm_to_m(self.length1)
            elif self.length2_name == "km":
                result = mm_to_km(self.length1)
            elif self.length2_name == "inch":
                result = mm_to_inch(self.length1)
            elif self.length2_name == "yard":
                result = mm_to_yard(self.length1)
            elif self.length2_name == "mile":
                result = mm_to_mile(self.length1)
            else:
                return f'INVALID CONVERT TO METRIC'
        elif self.length1_name == "cm":
            if self.length2_name == "mm":
                result = cm_to_mm(self.length1)
            elif self.length2_name == "m":
                result = cm_to_m(self.length1)
            elif self.length2_name == "km":
                result = cm_to_km(self.length1)
            elif self.length2_name == "inch":
                result = cm_to_inch(self.length1)
            elif self.length2_name == "yard":
                result = cm_to_yard(self.length1)
            elif self.length2_name == "mile":
                result = cm_to_mile(self.length1)
            else:
                return f'INVALID CONVERT TO METRIC'

        elif self.length1_name == "m":
            if self.length2_name == "mm":
                result = m_to_mm(self.length1)
            elif self.length2_name == "cm":
                result = m_to_cm(self.length1)
            elif self.length2_name == "km":
                result = m_to_km(self.length1)
            elif self.length2_name == "inch":
                result = m_to_inch(self.length1)
            elif self.length2_name == "yard":
                result = m_to_yard(self.length1)
            elif self.length2_name == "mile":
                result = m_to_mile(self.length1)
            else:
                return f'INVALID CONVERT TO METRIC'

        elif self.length1_name == "km":
            if self.length2_name == "mm":
                result = km_to_mm(self.length1)
            elif self.length2_name == "cm":
                result = km_to_cm(self.length1)
            elif self.length2_name == "m":
                result = km_to_m(self.length1)
            elif self.length2_name == "inch":
                result = km_to_inch(self.length1)
            elif self.length2_name == "yard":
                result = km_to_yard(self.length1)
            elif self.length2_name == "mile":
                result = km_to_mile(self.length1)
            else:
                return f'INVALID CONVERT TO METRIC'

        elif self.length1_name == "inch":
            if self.length2_name == "mm":
                result = inch_to_mm(self.length1)
            elif self.length2_name == "cm":
                result = inch_to_cm(self.length1)
            elif self.length2_name == "m":
                result = inch_to_m(self.length1)
            elif self.length2_name == "km":
                result = inch_to_km(self.length1)
            elif self.length2_name == "yard":
                result = inch_to_yard(self.length1)
            elif self.length2_name == "mile":
                result = inch_to_mile(self.length1)
            else:
                return f'INVALID CONVERT TO METRIC'

        elif self.length1_name == "yard":
            if self.length2_name == "mm":
                result = yard_to_mm(self.length1)
            elif self.length2_name == "cm":
                result = yard_to_cm(self.length1)
            elif self.length2_name == "m":
                result = yard_to_m(self.length1)
            elif self.length2_name == "km":
                result = yard_to_km(self.length1)
            elif self.length2_name == "inch":
                result = yard_to_inch(self.length1)
            elif self.length2_name == "mile":
                result = yard_to_mile(self.length1)
            else:
                return f'INVALID CONVERT TO METRIC'

        elif self.length1_name == "mile":
            if self.length2_name == "mm":
                result = mile_to_mm(self.length1)
            elif self.length2_name == "cm":
                result = mile_to_cm(self.length1)
            elif self.length2_name == "m":
                result = mile_to_km(self.length1)
            elif self.length2_name == "km":
                result = mile_to_inch(self.length1)
            elif self.length2_name == "inch":
                result = mile_to_inch(self.length1)
            elif self.length2_name == "yard":
                result = mile_to_yard(self.length1)
            else:
                return f'INVALID CONVERT TO METRIC'

        else:
            return f'INVALID CONVERT FROM METRIC'
        return self.summary(result)

    # ---------- Millimetres ------------------
    # MM to CM
    def mm_to_cm(self):
        return float(self.length1/10)
    # MM to M
    def mm_to_m(self):
        return float(self.length1/1000)
    # MM to KM
    def mm_to_km(self):
        return float(self.length1/1000000)
    # MM to Inches
    def mm_to_inch(self):
        return float(self.length1*0.0393701)
    # MM to Yards
    def mm_to_yard(self):
        return float(self.length1*0.00109361)
    # MM to Miles
    def mm_to_mile(self):
        return float(self.length1*6.2137e-7)
    
    # ---------- Centimeters -------------
    # CM to MM
    def cm_to_mm(self):
        return float(self.length1*10)
    # CM to M
    def cm_to_m(self):
        return float(self.length1/100)
    # CM to KM
    def cm_to_km(self):
        return float(self.length1/100000)
    # CM to Inch
    def cm_to_inch(self):
        return float(self.length1*0.3937)
    # CM to Yards
    def cm_to_yard(self):
        return float(self.length1*0.010936111996938)
    # CM to Miles
    def cm_to_mile(self):
        return float(self.length1*6.2137e-6)

    # --------------- Meters ----------------------
    # M to MM
    def m_to_mm(self):
        return float(self.length1*1000)
    # M to CM
    def m_to_cm(self):
        return float(self.length1*100)
    # M to KM
    def m_to_km(self):
        return float(self.length1/1000)
    # M to inches
    def m_to_inch(self):
        return float(self.length1*39.37006655984252)
    # M to Yrds
    def m_to_yard(self):
        return float(self.length1*1.0936)
    # M to Miles
    def m_to_mile(self):
        return float(self.length1*0.000621371)


    # ---------------Kilometers -------------------
# KM to MM
    def km_to_mm(self):
        return float(self.length1*1000000)
    # KM to CM
    def km_to_cm(self):
        return float(self.length1*100000)
    # KM to M
    def km_to_m(self):
        return float(self.length1*1000)
    # KM to inches
    def km_to_inch(self):
        return float(self.length1*39370.1)
    # KM to yards
    def km_to_yard(self):
        return float(self.length1*1093.613888889)
    # KM to Mile
    def km_to_mile(self):
        return float(self.length1*0.621371)

# ----------- Inch -------------------------
    # inch to MM
    def inch_to_mm(self):
        return (self.length1*25.3999512574)
    # inch to CM
    def inch_to_cm(self):
        return (self.length1*2.5399951257399999704)
    # inch to M
    def inch_to_m(self):
        return (self.length1*0.0254)
    # inch to KM
    def inch_to_km(self):
        return (self.length1*2.54e-5)
    # inch to yards
    def inch_to_yard(self):
        return (self.length1*0.0277778)
    # inch to MM
    def inch_to_mile(self):
        return (self.length1*1.5783e-5)

# ----------- Yard -----------------
    # Yards to MM
    def yard_to_mm(self):
        return float(self.length1*914.4)

    # Yards to CM
    def yard_to_cm(self):
        return float(self.length1*91.44)

    # Yrds to M
    def yard_to_m(self):
        return float(self.length1*0.9144)

    # Yards to KM
    def yard_to_km(self):
        return float(self.length1*0.0009144)

    # Yrds to inches
    def yard_to_inch(self):
        return float(self.length1*36)

    # Yards to Miles
    def yard_to_mile(self):
        return float(self.length1*0.000568182)

# ---------- Mile --------------------
    # Mile to MM
    def mile_to_mm(self):
        return float(self.length1*1.609e+6)

    # Mile to CM
    def mile_to_cm(self):
        return float(self.length1*160899.999998412)

    # Mile to M
    def mile_to_m(self):
        return float(self.length1*1609.3)

    # Mile to KM
    def mile_to_km(self):
        return float(self.length1*1.6093)

    # Mile to inches
    def mile_to_inch(self):
        return float(self.length1*63360)

    # Mile to Yard
    def mile_to_yard(self):
        return float(self.length1*1760)

```

```python

def convert_func(length1,length1_name,length2_name):
    if length1_name == "mm":
        if length2_name == "cm":
            result = float(length1/10)
        elif length2_name == "m":
            result = float(length1/1000)
        elif length2_name == "km":
            result = float(length1/1000000)
        elif length2_name == "inch":
            result = float(length1*0.0393701)
        elif length2_name == "yard":
            result = float(length1*0.00109361)
        elif length2_name == "mile":
            result = float(length1*6.2137e-7)
        else:
            return f'INVALID CONVERT TO METRIC'
    elif length1_name == "cm":
        if length2_name == "mm":
            result = float(length1*10)
        elif length2_name == "m":
            result = float(length1/100)
        elif length2_name == "km":
            result = float(length1/100000)
        elif length2_name == "inch":
            result = float(length1*0.3937)
        elif length2_name == "yard":
            result = float(length1*0.010936111996938)
        elif length2_name == "mile":
            result = float(length1*6.2137e-6)
        else:
            return f'INVALID CONVERT TO METRIC'

    elif length1_name == "m":
        if length2_name == "mm":
            result = float(length1*1000)
        elif length2_name == "cm":
            result = float(length1*100)
        elif length2_name == "km":
            result = float(length1/1000)
        elif length2_name == "inch":
            result = float(length1*39.37006655984252)
        elif length2_name == "yard":
            result = float(length1*1.0936)
        elif length2_name == "mile":
            result = float(length1*0.000621371)
        else:
            return f'INVALID CONVERT TO METRIC'

    elif length1_name == "km":
        if length2_name == "mm":
            result = float(length1*1000000)
        elif length2_name == "cm":
            result = float(length1*100000)
        elif length2_name == "m":
            result = float(length1*1000)
        elif length2_name == "inch":
            result = float(length1*39370.1)
        elif length2_name == "yard":
            result = float(length1*1093.613888889)
        elif length2_name == "mile":
            result = float(length1*0.621371)
        else:
            return f'INVALID CONVERT TO METRIC'

    elif length1_name == "inch":
        if length2_name == "mm":
            result = float(length1*25.3999512574)
        elif length2_name == "cm":
            result = float(length1*2.5399951257399999704)
        elif length2_name == "m":
            result = float(length1*0.0254)
        elif length2_name == "km":
            result = float(length1*2.54e-5)
        elif length2_name == "yard":
            result = float(length1*0.0277778)
        elif length2_name == "mile":
            result = float(length1*1.5783e-5)
        else:
            return f'INVALID CONVERT TO METRIC'

    elif length1_name == "yard":
        if length2_name == "mm":
            result = float(length1*914.4)
        elif length2_name == "cm":
            result = float(length1*91.44)
        elif length2_name == "m":
            result = float(length1*0.9144)
        elif length2_name == "km":
            result = float(length1*0.0009144)
        elif length2_name == "inch":
            result = float(length1*36)
        elif length2_name == "mile":
            result = float(length1*0.000568182)
        else:
            return f'INVALID CONVERT TO METRIC'

    elif length1_name == "mile":
        if length2_name == "mm":
            result = float(length1*1.609e+6)
        elif length2_name == "cm":
            result = float(length1*160899.999998412)
        elif length2_name == "m":
            result = float(length1*1609.3)
        elif length2_name == "km":
            result = float(length1*1.6093)
        elif length2_name == "inch":
            result = float(length1*63360)
        elif length2_name == "yard":
            result = float(length1*1760)
        else:
            return f'INVALID CONVERT TO METRIC'

    else:
        return f'INVALID CONVERT FROM METRIC'
    return f'{float(length1)}{length1_name} = {result}{length2_name}'
```
