```python
import plotly.graph_objects as go
df_stocks = px.data.stocks()
# EXAMPLE 1
px.line(df_stocks,x='date',y='AAPL', labels={'x':'Date','y':'Price'})
#EXAMPLE 2
px.line(df_stocks,x='date',y=['AAPL','GOOG'], labels={'x':'Date','y':'Price'},title="Apple vs Google") 
```
